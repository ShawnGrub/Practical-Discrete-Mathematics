### 2. Reading and Writing Text Files

```python
# Writing to a text file
with open('my_text_file.txt', 'w') as f:
    f.write('Hello, world!\n')
    f.write('This is a second line.')
print("Content written to my_text_file.txt")

# Reading from a text file
with open('my_text_file.txt', 'r') as f:
    content = f.read()
print("\nContent of my_text_file.txt:")
print(content)
```

### 3. Reading and Writing CSV Files (using pandas)

```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# Writing to a CSV file
df.to_csv('my_data.csv', index=False)
print("DataFrame written to my_data.csv")

# Reading from a CSV file
df_read = pd.read_csv('my_data.csv')
print("\nContent of my_data.csv:")
display(df_read)
```

### 4. Reading and Writing JSON Files

```python
import json

# Sample dictionary
my_dict = {
    'name': 'Python',
    'version': 3.9,
    'is_awesome': True,
    'features': ['readable', 'versatile', 'large community']
}

# Writing to a JSON file
with open('my_data.json', 'w') as f:
    json.dump(my_dict, f, indent=4)
print("Dictionary written to my_data.json")

# Reading from a JSON file
with open('my_data.json', 'r') as f:
    data_read = json.load(f)
print("\nContent of my_data.json:")
print(data_read)
print(f"Accessing a value: {data_read['version']}")
```
